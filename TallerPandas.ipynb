**5. ¿Cuál fue el ingreso total de los pedidos?**

```python
df.head(2)
```

```python
# Se debe crear una columna adicional con el totalprice = quantity * item_price
# Creo copio de la base para esto
aux = df.copy()
```

```python
aux.head(1)
```

```python
#Reemplazar el signo pesos para poder realizar la operacion matematica, float es un valor numerico decimal. 
#Creo una nueva columna para el float
aux['item_price_float'] = aux['item_price'].replace('[/$,]', '', regex=True).astype(float)
```

```python
aux.head()
```

```python
aux['total_price'] = aux['quantity'].astype(int) * aux['item_price_float'].astype(float)
```

```python
aux.head()
```

```python
aux2 = aux.groupby('order_id')['total_price'].sum()
aux2.head(8)
```
